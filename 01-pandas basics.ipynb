```python
# pandas.pydata.org
import pandas as pd

# Now us pd to read csv file countries.csv
# The result is a called a DataFrame = df   (df is a variable name)
df = pd.read_csv("countries.csv")

# Execute df to see the content of the file
df
```

```python
# Let's see the first 5 rows
# hold down shift and press tab to see the documentation
df.head()
```

```python
# Let's see the last 5 rows
df.tail()
```

```python
# how to get the life expectancy of every country
# df["life_expectancy"] is a series not a dataframe. thats why it looks different(ugly)
df["life_expectancy"]
```

```python
df["life_expectancy"]
```

```python
# to sort the dataframe by life_expectancy
df.sort_values("life_expectancy")
```

```python
# to sort the dataframe by life_expectancy in descending order
df.sort_values("life_expectancy", ascending=False)
```

```python
# to get two columns
# in this case i need two square brackets
df[["country", "life_expectancy"]]
```

```python
# give me the 10 contries with the highest life expectancy
# sort evertything by life expectancy and then give me the first 10
df[["country", "life_expectancy"]].sort_values("life_expectancy", ascending=False).head(10)
```

```python
# give me the first 10 contries than sort them by life expectancy
df.head(10).sort_values("life_expectancy", ascending=False)
```

```python
# filter the data
# by filtering use double df[df["column"] == "value"]
# goes through the dataframe and gives me true or false for every row
# df["continent"] == "Asia"

# the way to do it
df[df["continent"] == "Asia"]
```

```python
# comparison operators
# not df["life_expectancy"] > 80
df[df["life_expectancy"] > 80]

# Wrong
# df["life_expectancy"] > 80 and df["continent"] == "Asia"

# Correct
df[(df["life_expectancy"] > 80) & (df["continent"] == "Asia")]
```

```python
df['life_expectancy'].mean()
```

```python
df['life_expectancy'].median()
df['life_expectancy'].sum()
```

```python
# to get the number of rows like mean, max, min
df['life_expectancy'].describe()
```

```python
df[['life_expectancy', 'gdp']].corr()
```

```python
# calculate the per capita gdp
# gdp / population

df['gdp_per_capita'] = df['gdp'] / df['population']
df.head(2)
```

```python
# 1. we take a dataframe
# 2. filters
# 3. we group it by continent
# 4. columns
# 5. do math (we calculate the median)
df.groupby('continent')['life_expectancy'].median()
```

```python
df['continent'].value_counts()
```

## API?? or JSON??

```python
import requests

url = "http://api.weatherapi.com/v1/forecast.json?key=1deea713c0524e4abd621042220711&q=London&days=3&aqi=yes&alerts=no"
response = requests.get(url)
data = response.json()

# Put it in a DataFrame
# Data dont look good (because of lists and dictionaries)
pd.DataFrame(data['forecast']['forecastday'])

# Normalize the data with json_normalize
pd.json_normalize(data['forecast']['forecastday'])
```
